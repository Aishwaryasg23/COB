```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('dataset - netflix1 (2).csv')
df.head()
```

```python
print(plt.style.available)
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='type',data=df, palette='dark',hue='type',
             linewidth=1,edgecolor=sns.color_palette('dark',5)
             )
plt.style.use('dark_background')
plt.title('Count of show types on Netflix')
plt.xlabel('Show type')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(10,8))
plt.hist(df['release_year'],color='yellow')
sns.set_theme(style='darkgrid')
plt.title('Relese year of movies/TV Shows on Netflix')
plt.xlabel('Release year')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
#barplot of no. of content v/s rating
plt.figure(figsize=(10,8))
plt.style.use('fast')
sns.countplot(df['rating'],color='green')
plt.title('Number of content v/s Rating')
plt.xlabel('Number of content')
plt.ylabel(' Rating')
plt.grid(True)
plt.show()
```

```python
#Plot of Top 10 countries with highest content.
top_countries=df['country'].value_counts().nlargest(10)
plt.figure(figsize=(10,8))
plt.style.use('fast')
top_countries.plot(kind='bar',color='purple')
plt.title('Top 10 countries with highest content')
plt.xlabel("Name of the countries")
plt.ylabel("Count of Contents")
plt.legend()
plt.show()
```

```python
# Bar plot of Top 10years in which maximum movies are released.
top_years=df['release_year'].value_counts().nlargest(10)
plt.figure(figsize=(10,8))
plt.style.use('fast')
top_years.plot(kind='barh',color='pink')
plt.title("Top 10years in which maximum movies are released")
plt.ylabel('Year of Release')
plt.xlabel('Count')
plt.legend()
plt.show()
```

```python
#Top 15 genres watched on Netflix
top_list=df['listed_in'].value_counts().nlargest(15)
plt.figure(figsize=(10,8))
plt.style.use('fast')
top_list.plot(kind='bar',color='orange')
plt.title("Top 15 genres watched on Netflix")
plt.xlabel('Count')
plt.ylabel('Genres')
plt.legend()
plt.show()
```
