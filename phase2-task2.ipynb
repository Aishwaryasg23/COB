```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#1=train, 2-test dataset
# URL of the Google Sheets document
url1 = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vRTK2NvcndgPX41Czu6Ft2Ho_nE-z50BgTqdzwFW0rsJ2nvyNLe2DoIg1COzUbgw80oaRBjfy5-WtFk/pubhtml'
url2='https://docs.google.com/spreadsheets/d/e/2PACX-1vRyvZ7lknwiSghK9aen1SaTEYoN3JS40rrGLpcyrsVZy1tB2T4gn6Y3-cdzPUFCPMmmqREWefW3kl4_/pubhtml'
# Reading HTML tables
tables1 = pd.read_html(url1)
tables2= pd.read_html(url2)

#Accessing the data
df1 = tables1[0]  
df2=tables2[0]

# Display the Data
print(df1)
print(df2)
```

```python
#storing the test & train data into CSV file
df1.to_csv("train-dataset.csv", index=False, header=False)
df2.to_csv("test-dataset.csv", index=False, header=False)
train_df=pd.read_csv("train-dataset.csv")
test_df=pd.read_csv("test-dataset.csv")
```

```python
trimmedtest_df=test_df.dropna()
trimmedtest_df=test_df.drop(test_df.columns[0],axis=1)
trimmedtest_df
```

```python
trimmed_df=train_df.dropna()
trimmed_df
```

```python
trimmedtrain_df=trimmed_df.drop(train_df.columns[0], axis=1)
trimmedtrain_df
```

```python
#Accesing the column values from train dataset
x1=trimmedtrain_df.iloc[ : ,:1]
y1=trimmedtrain_df.iloc[ : ,1:2]
```

```python
#Accesing the column values from test dataset
x2=trimmedtest_df.iloc[ : ,:1]
y2=trimmedtest_df.iloc[ : ,1:2]
```

```python

from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(x1, y1)
```

```python
regressor.predict(x1)
```

```python

#visualizing the Train set results
%matplotlib inline
plt.figure(figsize=(8,6))
plt.scatter(x1, y1, color="black", marker='*')
plt.plot(x1, regressor.predict(x1), color="green")
plt.title("X V/S Y (Train Dataset)")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
regressor.fit(x2, y2)
regressor.predict(x2)
```

```python
#visualizing the Test set results

%matplotlib inline
plt.figure(figsize=(8,6))
plt.scatter(x2, y2, color="black", marker='*')
plt.plot(x2, regressor.predict(x2), color="green")
plt.title("X V/S Y (Test Dataset)")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
